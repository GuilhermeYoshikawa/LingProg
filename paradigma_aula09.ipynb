## 1. Implemente um gerador infinito de números primos.
Ex:

i = 0

while i < 10:

print(gera_primos())

i += 1

Este código imprime os 10 primeiros números primos, isto é: 1, 2, 3, 5, 7, 11, 13, 17,

19, 23

```python
def gera_primos():
    n = 1
    while True:
        if n == 0 or n == 1 or not(any(list(map(lambda i: n % i == 0, range(2, n))))):
            yield n
        n += 1

gerador = gera_primos()
gera_primo = lambda: gerador.send(None)

print("Este código imprime os 10 primeiros números primos, isto é: ")

i = 0
while i < 10:
    print(gera_primo())
    i += 1
```

## 4. Escreva um generator que crie um fluxo infinito de inteiros a partir de um valor inicial dado com passo 5.

Ex: gera_inteiros(7) = 12, 17, 22, 27, ...

```python
def gera_inteiros(inteiro):
    n = inteiro + 5
    yield n
    while True:
        n += 5
        yield n
        
gerador = gera_inteiros(7)
i = 0

print("gera_inteiros(7) =")
while i < 4:
    print(gerador.send(None))
    i += 1
```
