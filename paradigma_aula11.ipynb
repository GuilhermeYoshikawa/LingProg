## 1. Use map() para criar uma função que encontra o comprimento de cada palavra em uma frase (quebrada por espaços) e retorna os valores em um lista. A função deve ter como entrada uma string e como saída uma lista de inteiros.
Ex: word_lengths('How long are the words in this phrase') = [3, 4, 3, 3, 5, 2, 4, 6]

```python
word_lenght = lambda phrase: list(map(len, phrase.split()))
print("word_lengths('How long are the words in this phrase') =", word_lenght("How long are the words in the phrase"))
```

## 2. Use reduce() para pegar uma lista de inteiros e retornar o número que corresponde à combinação destes dígitos. Por exemplo, [1,2,3] corresponde a cento e vinte e três.
Ex: digits_to_num([3,4,3,2,1]) = 34321

```python
from functools import reduce

digits_to_num = lambda d: int(reduce(lambda x, y: str(x) + str(y), d))
print("digits_to_num([3,4,3,2,1]) =", digits_to_num([3, 4, 3, 2, 1]))
```

## 3. Use filter() para retornar as palavras de uma lista de palavras que comecem com uma determinada letra.
Ex:

l = ['hello','are','cat','dog','ham','hi','go','to','heart']

```python
filter_words = lambda words, letter: list(filter(lambda words: words.startswith(letter), words))  
l = ['hello','are','cat','dog','ham','hi','go','to','heart']

print("l = ['hello','are','cat','dog','ham','hi','go','to','heart']")
print("filter_words(l,'h') =", filter_words(l, 'h'))
```

## 5. Use itertools.cycle para criar um spinner (a antiga ampulheta do windows) no formato caractere. Use os caracteres \|/– para simular a rotação. A função spinner recebe como parâmetro o tempo de duração do spinner. Use time.time() para o tempo, sys.stdout.write() para escrever o caractere do spinner, sys.stdout.flush() ara disparar a thread e time.sleep() para reduzir a velocidade de rotação.

```python
from time import time
from time import sleep
from sys import stdout

def spinner(duration, time_betwen_ticks):

    target_time = time() + duration

    for tick in cycle(['\\', '|', '/', '–']):
        sleep(time_betwen_ticks)
        stdout.write(tick)
        stdout.flush()

        if time() >= target_time:
            break

spinner(14, 0.1)
```

## 6. Dado um dicionário cujas chaves sejam nomes de livros e seus respectivos valores sejam suas quantidades em estoque, use itertools.dropwhile para retornar livros com quantidade maior ou igual a 2.

```python
def quantidade_no_estoque(livros, quantidade):
    return list(it.dropwhile(lambda livro: livros[livro] >= quantidade, livros.keys()))

livros = {
    "Wonder": 10,
    "Do mil ao milhão": 5,
    "A culpa é das estrelas": 1,
}

print("Livros com quantidade menor ou igual a 2 =", quantidade_no_estoque(livros, 2))
```

## 7. Dada uma lista de alunos, quantos pares podem ser formados? Use itertools.combinations.

```python
def pares_formados(alunos):
    return len(list(it.combinations(alunos, r=2)))

alunos = ["Guilherme", "Erik", "Gustavo", "Felipe", "Henrique"]

print("Podem ser formados " + str(pares_formados(alunos)) + " pares de alunos")
```

## 8. Quantas palavras de 4 letras podem ser formadas a partir de uma lista de 7 letras. Use itertools.permutations.

```python
def palavras_formadas(lista):
    return len(list(it.permutations(lista, 4)))

print("Podem ser formadas " + str(palavras_formadas(['1', '2', '3', '4', '5', '6', '7'])) + " palavras de 4 letras a partir de uma lista de 7 letras.")
```

## 9. Considere um dicionário cuja chave são nomes de usuários e os respectivos valores são o meio de contato preferido (opções: e-mail, carta, whatsapp, telefone fixo, fax e celular). Crie uma função que retorna a quantidade de usuários que preferem e-mail. Use itertools.groupby()

```python
def prefere_email(usuarios):
    usuarios = {k: i for k, i in sorted(usuarios.items(), key = lambda item: item[1])}

    for key, group in it.groupby(usuarios.keys(), lambda user: usuarios[user]):
        if key == "e-mail":
            return len(list(group))
    return 0

usuarios = {
    "Guilherme" : "e-mail",
    "Erik": "carta",
    "Gustavo": "whatsapp",
    "Felipe": "telefone fixo",
    "Henrique": "fax",
}

print("Quantidade de usuários que preferem e-mail =", prefere_email(usuarios))
```
