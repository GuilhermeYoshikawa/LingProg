## Os exercícios abaixo são os mesmos do notebook da aula03. Implemente-os usando comprehensions, map, filter e lambdas. Utilizar maybe monads para fazer o tratamento de exceções quando necessário. 

```python
def maybe(fnc):
    def inner(*args):
        for a in args:
            if isinstance(a, Exception):
                return None
        try:
            return fnc(*args)
        except Exception as e:
            return None
    return inner
```

## 8. Mesma letra: Escreva uma função que receba uma string com duas palavras e retorne True se ambas palavras começarem com a mesma letra. 
Exemplo:

mesma_letra('Cão covarde') -> True

mesma_letra('Vira Lata') -> False

```python
def mesma_letra(palavra):
    palavras = palavra.lower().split(" ")
    return palavras[0][0] == palavras[1][0]

mesma_letra = maybe(mesma_letra)
print("mesma_letra('Cão covarde') ->", mesma_letra("Cão covarde"))
print("mesma_letra('Vira Lata') ->", mesma_letra("Vira Lata"))
```

## 9. Mestre Yoda: Dada uma sentença, a função deve retornar a sentença com as palavras na ordem reversa. Exemplo:
mestre_yoda('Eu estou em casa') --> 'casa em estou Eu'

mestre_yoda('Estamos prontos') --> 'prontos Estamos'

```python
def mestre_yoda(frase):
    return " ".join(list(reversed(frase.lower().split(" ")))).capitalize()

mestre_yoda = maybe(mestre_yoda)
print("mestre_yoda('Eu estou em casa') -->", mestre_yoda("Eu estou em casa"))
print("mestre_yoda('Estamos prontos') -->", mestre_yoda("Estamos prontos"))
```

## 10. Espião: Escreva uma função que receba uma lista de inteiros e retorne True se contém um 007 em ordem, mesmo que não contínuo. 
Exemplo:

espiao([1,2,4,0,0,7,5]) --> True

espiao([1,0,2,4,0,5,7]) --> True

espiao([1,7,2,4,0,5,0]) --> False

```python
def espiao(lista):
    primeiro = False
    segundo = False
    for i in lista:
        if segundo and i == 7:
            return True
        if not segundo and primeiro and i == 0:
            segundo = True
        if not segundo and i == 0:
            primeiro = True
    return False

espiao = maybe(espiao)
print("espiao([1,2,4,0,0,7,5]) -->", espiao([1,2,4,0,0,7,5])) 
print("espiao([1,0,2,4,0,5,7]) -->", espiao([1,0,2,4,0,5,7])) 
print("espiao([1,7,2,4,0,5,0]) -->", espiao([1,7,2,4,0,5,0]))
```
