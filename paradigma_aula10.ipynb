## Os exercícios abaixo são os mesmos do notebook da aula03. Implemente-os usando comprehensions, map, filter e lambdas. Utilizar maybe monads para fazer o tratamento de exceções quando necessário. 

```python
def maybe_monads(fnc):
    def inner(*args):
        for i in args:
            if isinstance(i, Exception):
                return i
            try:
                return fnc(*args)
            except Exception as ex:
                return ex
            return inner
```

## 1. Faça um programa que leia 5 números e informe o maior número.

```python
lista = []

for i in range(0,5):
    n = int(input("Insira o " + str(i + 1) + "° número: "))
    lista.append(n)
    
    maior = lista[0]
    for n in lista:
        if n > maior:
            maior = n
            
print("Maior número: " + str(maior))
```

## 2. Faça um programa que leia 5 números e informe a soma e a média dos números.

```python
lista = []

for i in range(0, 5):
    n = int(input("Insira o " + str(i + 1) + "° número: "))
    lista.append(n)
    
    s = 0
    for n in lista:
        s += n
        
print("Soma dos números: " + str(s))
print("Média dos números: " + (str(s / 5)))
```

## 4. Desenvolva um gerador de tabuada, capaz de gerar a tabuada de qualquer número inteiro entre 1 a 10. O usuário deve informar de qual numero ele deseja ver a tabuada. A saída deve ser conforme o exemplo abaixo: 

Tabuada de 5:

5 X 1 = 5

5 X 2 = 10

...

5 X 10 = 50

```python
multiplo = int(input("Qual número você deseja ver a tabuada (entre 1 a 10): "))
tabuada = (lambda n: print(str(multiplo) + " x " + str(n) + " = " + str(multiplo * n)))

list(map(tabuada, range(1,11)))
```

## 5. Faça um programa que calcule o fatorial de um número inteiro fornecido pelo usuário. 
Ex.: 5!=5.4.3.2.1=120

```python
n = int(input("Insira o número que deseja descobrir a fatorial: "))

fatorial = 1
for i in range(1, n + 1):
    fatorial *= i

print("Fatorial =", fatorial)
```

Quantidade de Parcelas % de Juros sobre o valor inicial da dívida

1 0

12 25

Exemplo de saída do programa:

Valor da Dívida Valor dos Juros Quantidade de Parcelas Valor da Parcela

R$ 1.000,00 0 1 R$ 1.000,00

```python
tabela = [
    {"parcelas": 1,  "juros": 0.00},
    {"parcelas": 3,  "juros": 0.10},
    {"parcelas": 6,  "juros": 0.15},
    {"parcelas": 9,  "juros": 0.20},
    {"parcelas": 12, "juros": 0.25}
]

divida = float(input("Dívida: R$")) 
print("Valor da Dívida  Valor dos Juros  Quantidade de Parcelas  Valor da Parcela")
programa = (lambda coluna: print("R$ {:.2f}          R$ {:.2f}          {:d}                       R$ {:.2f}".format((divida * coluna["juros"] + divida), divida * coluna["juros"], coluna["parcelas"], (divida * coluna["juros"] + divida)/coluna["parcelas"])))

list(map(programa, tabela))
```
