## 1. Crie um programa que recebe uma lista de números e

```python
list = []
maior = 0
menor = 0
media = 0
soma_dos_negativos = 0

for c in range(0, 4):
    list.append(int(input('Digite um valor: ')))
    if c == 0:
        maior = menor = list[c]
    else:
        if list[c] > maior:
            maior = list[c]
            
media = sum(list) / len(list)

soma_dos_negativos = sum(filter(lambda c: c < 0, list))

print('=-' * 30)
print(f'Os valores digitados são: {list}')
print('=-' * 30)
print(f'O maior número inserido foi: {maior}')
print('=-' * 30)
print(f'A média da lista é: {media}')
print('=-' * 30)
print(f'Soma dos elementos com o valor negativo: {soma_dos_negativos}')
```

## 4. Faça um programa que percorre uma lista com o seguinte formato: [['Brasil', 'Italia', [10, 9]], ['Brasil', 'Espanha', [5, 7]], ['Italia', 'Espanha', [7,8]]]. Essa lista indica o número de faltas que cada time fez em cada jogo. Na lista acima, no jogo entre Brasil e Itália, o Brasil fez 10 faltas e a Itália fez 9. O programa deve imprimir na tela:

- o total de faltas do campeonato

- o time que fez mais faltas

- o time que fez menos faltas

```python
times = [['Brasil', 'Italia', [10, 9]], ['Brasil', 'Espanha', [5, 7]], ['Italia', 'Espanha', [7,8]]]

brasil_faltas = times[0][2][0] + times[1][2][0]
italia_faltas = times[0][2][1] + times[2][2][0]
espanha_faltas = times[1][2][1] + times[2][2][1]
faltas = brasil_faltas + italia_faltas + espanha_faltas

print("O total de faltas do campeonato: ", faltas)

if brasil_faltas > italia_faltas and brasil_faltas > espanha_faltas:
    print("Cometeu mais faltas: Brasil")
if italia_faltas > brasil_faltas and italia_faltas > espanha_faltas:
    print("O time que fez mais faltas: Itália")
else: 
    print("O time que fez mais faltas: França")
    
if brasil_faltas < italia_faltas and brasil_faltas < espanha_faltas:
    print("O time que fez mais faltas: Brasil")
if italia_faltas < brasil_faltas and italia_faltas < espanha_faltas:
    print("O time que fez mais faltas: Itália")
else: 
    print("O time que fez mais faltas: França")
```

## 6. Escreva um programa que lê duas notas de vários alunos e armazena tais notas em um dicionário, onde a chave é o nome do aluno. A entrada de dados deve terminar quando for lida uma string vazia como nome. Escreva uma função que retorna a média do aluno, dado seu nome.

```python
dicionario = {}
i = 0

aluno = input("Aluno: ")
while aluno != "":
    dicionario[aluno] = [
        float(input("Nota 1: ")),
        float(input("Nota 2: "))
    ]
    aluno = input("Aluno: ")

def media(aluno):
    return (dicionario[aluno][0] + dicionario[aluno][1]) / 2

media(input("Insira o nome do aluno o qual deseja ver a média: "))
```
